```
import json

#put all counts in a list 
snippet_size_counts = []
cleaned_snippet_size_counts = []

with open('final-combined-dataset.json') as json_file:
    final_combined_list = json.load(json_file)

    max_snippet_size = 0
    max_cleaned_snippet_size = 0
    max_intent_size = 0
    

    for i in range(0, len(final_combined_list)):
        #print(str(final_combined_list[i]))
        #print()
        curr_dict = final_combined_list[i]
        cleaned_snippet_size_counts.append(len(curr_dict['clean_snippet']))
        snippet_size_counts.append(len(curr_dict['snippet']))


        if len(curr_dict['snippet']) > max_snippet_size:
            max_snippet_size = len(curr_dict['snippet'])


        if len(curr_dict['clean_snippet']) > max_cleaned_snippet_size:
            max_cleaned_snippet_size = len(curr_dict['clean_snippet'])

        if len(curr_dict['intent']) > max_intent_size:
            max_intent_size = len(curr_dict['intent'])


        #print("snippet: " + str(curr_dict['clean_snippet']))
        #print(len(curr_dict['clean_snippet']))

        #print()
        #print("intent: " + str(curr_dict['intent']))
        #print(len(curr_dict['intent']))

        #print()
        #print()

    print("max intent size: " + str(max_intent_size))
    print("max cleaned snippet size: " + str(max_cleaned_snippet_size))
    print("max snippet size: " + str(max_snippet_size))
    #print("snippet size counts list: " + str(snippet_size_counts))

json_file.close()




#main()
```
