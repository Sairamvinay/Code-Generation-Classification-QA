```python
import json
with open('curated_mined-0.5.json') as f:
    mined = json.load(f)
    
with open('conala-corpus/conala-train.json') as f:
    train = json.load(f)
    
with open('conala-corpus/conala-test.json') as f:
    test = json.load(f)
```

```python
len(train)
```

```python
len(test)
```

```python
mined_intents = {}
questions_answers = {}
for qs in mined:
    if qs[0]['intent'] in mined_intents:
        mined_intents[qs[0]['intent']] += 1
    else:
        mined_intents[qs[0]['intent']] = 1
    if qs[0]['intent'] in questions_answers:
        questions_answers[qs[0]['intent']].append(qs)
    else:
        questions_answers[qs[0]['intent']] = qs
```

```python
print(len(questions_answers))
print(len(questions_answers.items()))
local_c = 0
for q in questions_answers:
    local_c += len(questions_answers[q])
local_c
```

```python
len(mined_intents)
```

```python
train_intents = {}
for a in train:
    if a['intent'] in train_intents:
        train_intents[a['intent']] += 1
    else:
        train_intents[a['intent']] = 1
    if a['intent'] in questions_answers:
        questions_answers[a['intent']].append(a)
    else:
        questions_answers[a['intent']] = [a]
```

```python
len(train_intents)
local_c = 0
for q in questions_answers:
    local_c += len(questions_answers[q])
local_c
```

```python
test_intents = {}
for a in test:
    if a['intent'] in test_intents:
        test_intents[a['intent']] += 1
    else:
        test_intents[a['intent']] = 1
    if a['intent'] in questions_answers:
        questions_answers[a['intent']].append(a)
    else:
        questions_answers[a['intent']] = [a]
```

```python
len(test_intents)
local_c = 0
for q in questions_answers:
    local_c += len(questions_answers[q])
local_c
```

```python
train_mined_overlap_count = 0
for intent in train_intents:
    if intent in mined_intents.keys():
        train_mined_overlap_count += 1
print(train_mined_overlap_count)
```

```python
test_mined_overlap_count = 0
for intent in test_intents:
    if intent in mined_intents.keys():
        test_mined_overlap_count += 1
print(test_mined_overlap_count)
```

```python
total_answer_count = 0
total_intents = []
for intent in [*mined_intents.keys(), *train_intents.keys(), *test_intents.keys()]:
    if intent not in total_intents:
        total_intents.append(intent)
        if intent in mined_intents:
            total_answer_count += mined_intents[intent]
        if intent in train_intents:
            total_answer_count += train_intents[intent]
        if intent in test_intents:
            total_answer_count += test_intents[intent]
```

```python
for question in questions_answers:
    q_id = None
    for answer in questions_answers[question]:
        if q_id is None:
            q_id = answer['question_id']
        else:
            if answer['question_id'] != q_id:
                print('question', question)
```

```python
with open('conala-all.jsonl', 'w+') as output_file:
    for key in questions_answers.keys():
        for a in questions_answers[key]:
            output_file.write('{}\n'.format(a))
#         json.dump(questions_answers, output_file, indent=4)
```

```python
# Create set of answers for each question (intent)
# Take answers and use them to take from questions_answers to create new
all_qs = []
for key in questions_answers.keys():
    s = set()
    for a in questions_answers[key]:
        s = s.union([a['snippet']])
    print(s)
    answers = []
    for el in s:
        print(el)
        for a in questions_answers[key]:
            if el == a['snippet']:
                answers.append(a)
                break
    all_qs += answers
```
